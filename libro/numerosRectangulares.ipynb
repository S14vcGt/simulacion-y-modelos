# Rectangular Numbers generation

Rectangular numbers generation produces *partialy random numbers* using a sucesion that follows a uniformal distribution, and are used to generate random variables in simulation experiments.
In this case, I'm going to use two linear congruential generators for the task. Linear congruential generators produce a secuence of pseudo random numbers in which the next pseudo random number is determined from the last generated number, that is, the pseudo random number $X_{n+1}$ is derived from the pseudo random number $X_n$.

## Mixed Congruential generator

In the particular case of the Mixed congruential generator the relation of recurrency is as follows: $$X_{n+1}= a(X_n+c)\mod m$$ 

Where:
$X_0$ = the seed ($X_0 > 0$)

$a$ = the multiplier ($a > 0$)

$c$ = additive constant ($c > 0$)

$m$ = the modulus ($m > X_0, a, c$)

The quantitie of possible numbers generated by this formula is, in a case where the parameters were rightfully chosen, equivalent to $m$. When this is like that, we say that we got a full period, and is not always the case. If we want to always get a full period we must follow a set of rules for chosing the parameters of the generator:

1. $m$ must be the greatest prime number that is minor than $p^d$
    - where: 
        - $p$: the base of the system ( binary, decimal, hexadecimal)
        - $d$: the number of bits of a computer instruction in the system( 64 bits, 32 bits)
2. $a$
    - integer
    - $(a - 1) \mod 4 = 0$ if 4 is a factor of $m$
    - $(a - 1) \mod b = 0$ if $b$ is a prime factor of $m$
3. $c$
    - integer
    - $c \mod 8 =5$ if is binary system
    - $c \mod 200 = 21$ if is decimal system
    - $c$ must be odd and relatively prime to $m$ 
4. $X_0$ doesn't care what value get


```python
# so now lets use the formulas and generate some rectangular numbers

import random as rd


def sacarB(numero): # its a greedy algo bc is a lot of operations
    factores = []
    for i in range(1, numero + 1):
        if numero % i == 0:
            factores.append(i)
    
    for posible_b in factores:
        for i in range(2, int(posible_b**0.5) + 1):
            if posible_b % i != 0: 
                return posible_b



def sacarA(m):
    b = sacarB(m)
    print(b)
    seguro_se_rompe = [i for i in range(3,m)]

    for posible_a in seguro_se_rompe:
        if (posible_a - 1) % b == 0:
            return posible_a
        
def sacarC():
    return 8 * rd.randint(45, 98) + 5

def mixed_generator(m,a,c,x0):
    return (a*(x0+c)) % m

# my system is 64 bits so in python is 2**63 - 1      
m = 2**15 - 1 # yeah, just a random number, because using a real proper number, would take like hours to give an output

a = 8#sacarA(m)

c = 373#sacarC()

x0= 81#rd.randint(1,89)

n = 1
print(f"seed: {x0}, multiplier: {a}, additive constant: {c}")
for _ in range (10):
    actual = mixed_generator(m,a,c,x0)
    print(f" rectangular number {n}: {actual}")
    x0 = actual
    n+=1

```
